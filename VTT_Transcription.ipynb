```python
#@title Download Audio from YouTube
video_url = 'https://www.youtube.com/watch?v=zPtiwNk2LOU'
!rm -rfv ./audio ./video *.7z
!mkdir -p audio video
!yt-dlp -f bestaudio '{video_url}' -o './audio/%(title)s_%(id)s_%(epoch)d.%(ext)s'
!yt-dlp -f bestvideo+bestaudio '{video_url}' -o './video/%(title)s_%(id)s_%(epoch)d.%(ext)s'
```

```python
#@title Transcribe Audio to VTT
import os
source_audio = os.path.abspath(os.path.join('audio', os.listdir('audio')[0]))
!ffmpeg -i "{source_audio}" -ar 16000 -ac 1 -c:a pcm_s16le 'output.wav'
!whisper 'output.wav' --model large-v3 -f vtt --language en -o './audio'
```

```python
#@title Upload to Drive
import os
from google.colab import drive
from datetime import datetime

drive.flush_and_unmount()
drive.mount('./drive')

archive = datetime.utcnow().strftime('%Y-%m-%d_%H-%M-%S')
archive = f'{archive}.7z'
!7z a -t7z -mx=9 {archive} './audio' './video'

destination_path = './drive/MyDrive/transcribed'
!mkdir -p {destination_path}
!cp -rfv {archive} {destination_path}
```
